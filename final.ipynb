```python
def univariate_plots(column, data=df):
    if data[column].dtype not in ['int64', 'float64']:
        f, axes = plt.subplots(1,1,figsize=(15,5))
        sns.countplot(x=column, data = data)
        plt.xticks(rotation=90)
        plt.suptitle(column,fontsize=20)
        plt.show()
    else:
        g = sns.FacetGrid(data, margin_titles=True, aspect=4, height=3)
        g.map(plt.hist,column,bins=100)
        plt.show()
    plt.show()
```

1. Linear Regression

```python
categorical_columns = []

for column in df_original.columns:
    if df_original[column].dtype == 'object':
        categorical_columns.append(column)
        
df_original = pd.get_dummies(df_original,columns=categorical_columns, dtype=int, drop_first=True)
df_original.fillna(0, inplace=True)

y = df_original['price']
X = df_original.drop(labels = ['price'], axis = 1)

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42, shuffle=True)

model = LinearRegression().fit(X_train, y_train)
y_pred = model.predict(X_test)

from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('r2_score:', metrics.r2_score (y_test, y_pred))
```

4 DIFFERENT MODELS

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso

categorical_columns = []

for column in df.columns:
    if df[column].dtype == 'object':
        categorical_columns.append(column)
        
df = pd.get_dummies(df,columns=categorical_columns, dtype=int, drop_first=True)
df.fillna(0, inplace=True)

y = df['price']
X = df.drop(labels = ['price'], axis = 1)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42, shuffle=True)

models = [DecisionTreeRegressor(), LinearRegression(), Ridge(),  Lasso()]

for model in models:
    
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    from sklearn import metrics
    print('Model:', model)
    print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
    print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
    print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
    print('r2_score:', metrics.r2_score (y_test, y_pred))
    print('-------------------------------------')
```

- It imports necessary modules from the scikit-learn library (sklearn.tree.DecisionTreeRegressor, sklearn.linear_model.LinearRegression, sklearn.linear_model.Ridge, sklearn.linear_model.Lasso).
- It initializes an empty list categorical_columns to store the names of categorical columns.
- It iterates through each column in the DataFrame df. If the data type of a column is 'object', it appends the column name to categorical_columns.
- It converts categorical variables into dummy/indicator variables using one-hot encoding via pd.get_dummies() method and assigns the result back to the DataFrame df.
- It fills any missing values in the DataFrame with 0.
- It separates the target variable 'price' from the features, assigning the features to X and the target variable to y.
- It splits the dataset into training and testing sets using train_test_split() method.
- It initializes a list models containing instances of different regression models: DecisionTreeRegressor, LinearRegression, Ridge, and Lasso.
- It iterates through each model in models, fits the model to the training data, predicts the target variable on the testing data, and then evaluates the performance of the model using various regression metrics such as Mean Absolute Error, Mean Squared Error, Root Mean Squared Error, and r2_score.
- Finally, it prints out the evaluation metrics for each model.

- Decision Tree Regressor:
Decision Tree Regressor is a supervised learning algorithm used for regression tasks.
It works by recursively splitting the dataset into subsets based on the value of a feature.
The splits are made in a way that minimizes the variance of the target variable within each subset.
Decision trees are prone to overfitting if not pruned properly, but they can capture complex relationships in the data.
- Linear Regression:
Linear Regression is a simple and widely used regression technique.
It assumes a linear relationship between the independent variables and the dependent variable.
Linear Regression tries to find the best-fitting straight line through the data points.
It is sensitive to outliers and multicollinearity but easy to interpret and computationally efficient.
- Ridge Regression:
Ridge Regression is a regularization technique that adds a penalty term to the linear regression cost function.
The penalty term is the L2 norm (sum of the squared coefficients) multiplied by a regularization parameter (alpha).
Ridge Regression shrinks the coefficients towards zero, which helps in reducing overfitting.
It is particularly useful when there is multicollinearity among the independent variables.
- Lasso Regression:
Lasso Regression is another regularization technique similar to Ridge Regression.
Instead of the L2 norm penalty, Lasso uses the L1 norm penalty (sum of the absolute coefficients) multiplied by the regularization parameter (alpha).
Lasso Regression not only helps in reducing overfitting but also performs feature selection by driving some coefficients to exactly zero.
It is useful when dealing with high-dimensional datasets with many irrelevant features.
